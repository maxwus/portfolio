```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
html_text = requests.get('https://www.nejlevnejsi-knihy.cz/vyhledavani.html?q=one+piece+omnibus&OP=1').text
```

```python
soup = BeautifulSoup(html_text, 'lxml')
OP_omnibuses = soup.find_all('div', class_ = 'LST_inf')
OP_prices = soup.find_all('div', class_ = "LST_buy")
```

```python
OP_dict_list= []
OP_unavailable = []
```

```python
#add divisio to name adn part, and avilability

for omnibus,price in zip(OP_omnibuses, OP_prices):      
    title = omnibus.find('h3').text
    price = price.find('p' ).text
    OP_dict_list.append({'title': title, 'price': price[:-3]})
        
    OP_df = pd.DataFrame.from_records(OP_dict_list)
    # print(title,price)
```
